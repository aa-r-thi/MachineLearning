# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing the dataset

```python
path = "/content/drive/MyDrive/Machine learning/PolynomialLinearRegression.csv"
dt = pd.read_csv(path)
X = dt.iloc[:,0:1].values
Y = dt.iloc[:,1:2].values
```

# Checking Null Values

```python
dt.isnull().sum()
```

# Training the linearRegrssion data on whole data

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,Y)
```

# Training the polynomial regression data on whole data



```python

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

pf = PolynomialFeatures(degree=6)
X_poly= pf.fit_transform(X)  # Assign the transformed array to X_poly

lr1 = LinearRegression()
lr1.fit(X_poly, Y)
```

# Visualize the Linear regression result

```python
plt.scatter(X,Y,color = "red")
plt.plot(X,lr.predict(X))
plt.xlabel("Age")
plt.ylabel("RBC")
plt.title("RBC count")
plt.show()
```

# Visualize the polynomial regression result

```python
plt.scatter(X,Y,color = "red")
plt.plot(X,lr1.predict(X_poly))
plt.xlabel("Age")
plt.ylabel("RBC")
plt.title("RBC count")
plt.show()
```

# Result prediction using linear regression

```python
lr.predict([[6.5]])
```

# Result prediction using polynomial regression

```python
lr1.predict(pf.fit_transform([[10]]))
```
